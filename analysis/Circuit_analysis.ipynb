```python
df = pd.read_csv('/content/drive/MyDrive/SERS_work/QuantumBenchmarks_Entire_Dataset - Sheet1.csv')


df['Gate Reduction'] = df['Two-Qubit Gate Count Before Optimization'] - df['Two-Qubit Gate Count After Optimization']

# List to store the selected QASM file names
selected_qasm_files = []

# Iterate through the DataFrame in steps of 3
for i in range(2, len(df), 3):  # Start at index 2, then 5, 8, etc.
    if i - 1 >= 0:  # Ensure the second row exists
        second_reduction = df.loc[i - 1, 'Gate Reduction']  # Row before
        third_reduction = df.loc[i, 'Gate Reduction']  # Current row

        if third_reduction > second_reduction:  # Compare reductions
            selected_qasm_files.append(df.loc[i, 'Circuit QASM File Name Before Optimization'])

# Create the new DataFrame with the selected QASM file names
filtered_df = pd.DataFrame(selected_qasm_files, columns=['Circuit QASM File Name Before Optimization'])
filtered_df.to_csv('/content/drive/My Drive/differing_circuits.csv', index=False)
```

```python
df = pd.read_csv('/content/drive/MyDrive/SERS_work/QuantumBenchmarks_Entire_Dataset - Sheet1.csv')
plt.figure(figsize=(12, 8))

df = df[df['Compilation Time (seconds)'] > 0]
df = df[df['Framework'] == 'BQSkit']

bins = np.arange(0,51,2)

def remove_high_outliers(group):
    Q1 = group['Compilation Time (seconds)'].quantile(0.25)
    Q3 = group['Compilation Time (seconds)'].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    return group[group['Compilation Time (seconds)'] <= upper_bound]


df_filtered = df.groupby('Circuit Qubit Count').apply(remove_high_outliers).reset_index(drop=True)

avg_times = df_filtered.groupby('Circuit Qubit Count')['Compilation Time (seconds)'].mean()


plt.hist(avg_times.index, weights=avg_times.values, bins=len(avg_times), edgecolor="black", color='royalblue')
plt.xticks(bins)

plt.xlabel('Circuit Qubit Count', fontweight='bold')
plt.ylabel('Compilation Time (Seconds)', fontweight='bold')
plt.title('Qiskit', fontweight='bold')
plt.show()
```
